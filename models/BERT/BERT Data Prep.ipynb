```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
train_test_ratio = 0.80
train_valid_ratio = 0.50

first_n_words = 200
```

```python
raw_data_path = source_folder + '/dataset.csv'
```

```python
def trim_string(x):

    x = x.split(maxsplit=first_n_words)
    x = ' '.join(x[:first_n_words])

    return x
```

```python
# Read raw data
df_raw = pd.read_csv(raw_data_path)

# Prepare columns
df_raw['humor'] = (df_raw['humor'] == True).astype('int')

# Drop rows with empty text
df_raw.drop( df_raw[df_raw.text.str.len() < 5].index, inplace=True)

# Trim text and titletext to first_n_words
df_raw['text'] = df_raw['text'].apply(trim_string)

assert len(df_raw) == 200000

# Train-test split
df_train, df_full_test = train_test_split(df_raw, train_size = train_test_ratio)

# Train-valid split
df_test, df_valid = train_test_split(df_full_test, train_size = train_valid_ratio)

# Write preprocessed data
df_train.to_csv(destination_folder + '/train.csv', index=False)
df_valid.to_csv(destination_folder + '/valid.csv', index=False)
df_test.to_csv(destination_folder + '/test.csv', index=False)
```
