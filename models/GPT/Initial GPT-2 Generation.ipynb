```python
import pandas
from tqdm import tqdm
from os.path import exists
from sklearn.model_selection import train_test_split
```

```python
df = pandas.read_csv(root_path + '/Data/dataset.csv')
```

```python
if(exists(root_path + '/Data/train.txt')):
  print('already exists')
else:
  dataset = []
  for i, row in tqdm(df.iterrows()):
    if (row['humor']):
      dataset.append('<|joke|>' + row['text'].strip())
  train, eval = train_test_split(dataset, train_size=.9, random_state=2021)

  with open(root_path + '/Data/train.txt', 'w') as file_handle:
    file_handle.write("<|endoftext|>".join(train))

  with open(root_path + '/Data/eval.txt', 'w') as file_handle:
    file_handle.write("<|endoftext|>".join(eval))

    print("training size:" + str(len(train)))
    print("Evaluation size: " + str(len(eval)))
```

```python
# setup imports to use the model
from transformers import TFGPT2LMHeadModel
from transformers import GPT2Tokenizer

model = TFGPT2LMHeadModel.from_pretrained(root_path + '/output', from_pt=True)
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
input_ids = tokenizer.encode("<|joke|>", return_tensors='tf')
```

```python
generated_text_samples = model.generate(
    input_ids, 
    max_length=150,  
    num_return_sequences=100,
    no_repeat_ngram_size=2,
    repetition_penalty=1.5,
    top_p=0.92,
    temperature=.85,
    do_sample=True,
    top_k=125,
    early_stopping=True
)
```

```python
#Print output for each sequence generated above
for i, output in enumerate(generated_text_samples):
  print("{}: {}".format(i,tokenizer.decode(output, skip_special_tokens=True)))
  print()
```

```python
#Print output for each sequence generated above
output_seq = []
for i, output in enumerate(generated_text_samples):
  output_seq.append('"' + tokenizer.decode(output, skip_special_tokens=True)[8:] + '"')
print(','.join(output_seq))
```
