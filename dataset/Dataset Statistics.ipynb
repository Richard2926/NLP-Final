```python
import nltk
nltk.download('punkt')
```

```python
from collections import Counter

token_counter = Counter()
total_vocab_count = 0
total_char_length = 0
total_samples = 0

for i, row in tqdm(df.iterrows()):
  tokens = nltk.word_tokenize(row['text'])
  token_counter.update(tokens)
  total_vocab_count += len(tokens)
  total_char_length += len(row['text'])
  total_samples += 1
```

```python
print('average words per phrase: ' + str(total_vocab_count / total_samples))
print('average chars per phrase: ' + str(total_char_length / total_samples))
print('total samples: ' + str(total_samples))
print('token count: ' + str(total_vocab_count))
print('vocab size: ' + str(len(token_counter)))
print('most popular words: ' + str(token_counter.most_common(100)))
```

```python
joke_token_counter, headline_token_counter = Counter(), Counter()
joke_total_vocab_count, headline_total_vocab_count = 0, 0
joke_total_char_length, headline_total_char_length = 0, 0
joke_total_samples, headline_total_samples = 0, 0

for i, row in tqdm(df.iterrows()):
  tokens = nltk.word_tokenize(row['text'])
  if (row['humor']):
    joke_token_counter.update(tokens)
    joke_total_vocab_count += len(tokens)
    joke_total_char_length += len(row['text'])
    joke_total_samples += 1
  else:
    headline_token_counter.update(tokens)
    headline_total_vocab_count += len(tokens)
    headline_total_char_length += len(row['text'])
    headline_total_samples += 1
```

```python
print('[joke] average words per phrase: ' + str(joke_total_vocab_count / joke_total_samples))
print('[joke] average chars per phrase: ' + str(joke_total_char_length / joke_total_samples))
print('[joke] total samples: ' + str(joke_total_samples))
print('[joke] token count: ' + str(joke_total_vocab_count))
print('[joke] vocab size: ' + str(len(joke_token_counter)))
print('[joke] most popular words: ' + str(joke_token_counter.most_common(100)))

print('---------------------------------------------')

print('[headline] average words per phrase: ' + str(headline_total_vocab_count / headline_total_samples))
print('[headline] average chars per phrase: ' + str(headline_total_char_length / headline_total_samples))
print('[headline] total samples: ' + str(headline_total_samples))
print('[headline] token count: ' + str(headline_total_vocab_count))
print('[headline] vocab size: ' + str(len(headline_token_counter)))
print('[headline] most popular words: ' + str(headline_token_counter.most_common(100)))
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
stop_words.add('.')
stop_words.add(',')
stop_words.add('?')
stop_words.add('\'s')
stop_words.add('(')
stop_words.add(')')
stop_words.add(':')
stop_words.add('n\'t')
stop_words.add('...')
stop_words.add('!')
stop_words.add('\'')
stop_words.add('\'m')
stop_words.add('\'re')
stop_words.add('’')
```

```python
def top10NoStop(stop_words, counter):
  count = 0
  top10 = []
  for word in counter.most_common(100):
    if word[0].lower() in stop_words:
      continue
    else:
      top10.append(word[0])
      count += 1
    if count == 10:
      break
  return top10
```

```python
print('Overall: ' + ','.join(top10NoStop(stop_words, token_counter)))
print('Jokes: ' + ','.join(top10NoStop(stop_words, joke_token_counter)))
print('Headline: ' + ','.join(top10NoStop(stop_words, headline_token_counter)))
```
